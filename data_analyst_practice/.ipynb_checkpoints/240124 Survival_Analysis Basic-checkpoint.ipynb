## 생존 함수

생존 함수를 그려보자. 기본적으로 여기서는 Kaplan-Meier estimation이라는 것을 사용해야 한다.

Link : https://www.kaggle.com/datasets/blastchar/telco-customer-churn

```python
df = pd.read_csv('C:/Users/USER/Desktop/TIL_new/data_analyst_practice/telco_customer_churn.csv')
```

```python
df.head(5)
```

변수별 영향도를 보기 이전에 근본적으로 생존함수, 즉 이 서비스를 이용했을 때 유저가 언제 이탈할지를 보는 생존함수를 그려보는 것이 목표이다.

```python
df.info()
```

```python
df.tenure.unique()
```

```python
df['is_churn'] = df.Churn.apply(lambda x: True if x == 'Yes' else False)
```

```python
from lifelines import KaplanMeierFitter

kmf = KaplanMeierFitter()
kmf.fit(df["tenure"], df["is_churn"])

plot = kmf.plot_survival_function()
plot.set_xlabel('time (month)')
plot.set_ylabel('survival function')
plot
```

2년 정도가 경과했을 때, 약 20% 정도의 유저가 churn, 즉 80%만이 생존했음을 알 수 있다.

## 집단간 비교

두 그룹의 생존 함수를 비교하여 유의미한 차이가 있는지 볼 수 있고, 시각화로 두 그룹의 생존함수 비교도 가능하다.

이를 고려하는 이유는 A/B 테스트에서의 이탈 방지 평가 등에도 활용 가능하고, 이탈이 취약한 집단을 찾는데에도 유의미할 것이다.

### Example
남성과 여성의 차이를 비교하자

```python
df_male = df.loc[df.gender == 'Male']
df_female = df.loc[df.gender == 'Female']
```

```python
kmf = KaplanMeierFitter()
kmf.fit(df_male['tenure'], df_male['is_churn'], label="male")
ax_kmf = kmf.plot()
kmf.fit(df_female['tenure'], df_female['is_churn'], label="female")
ax_kmf = kmf.plot(ax=ax_kmf)

ax_kmf.set_xlabel('time (month)')
ax_kmf.set_ylabel('survival function')
ax_kmf
```

그래프만 보면, 여성의 생존율이 살짝 낮아보이기는 한데, 문제는 이것이 통계적으로 유의미한 차이인지는 알 수 없다.
통계적 유의성을 확인하는 방법으로 logrank-test를 사용한다.

- 귀무가설 : 두 집단에 유의미한 차이가 없다.
- 대립가설 : 두 집단에 유의미한 차이가 존재한다.

```python
from lifelines.statistics import logrank_test
logrank_test(df_male["tenure"], df_female["tenure"], df_male["is_churn"], df_female["is_churn"]).p_value
```

p-value가 0.4684정도이다. 유의수준 5% 하에서, 귀무가설을 기각할 수 없다.
따라서, 남성과 여성 사이에 유의미한 이탈율 차이는 존재한다고 볼 수 없다.

## 이탈예측 모델

문제는, 우리는 각 변수를 유의미하게 사용하지 않았다는 점이다. 변수를 유의미하게 활용하여 이탈을 예측할 수 없을까?
정확히는 변수별로, 이탈에 대해 영향도가 다를 것인데 이를 고려하고 싶다는 의미이다.
